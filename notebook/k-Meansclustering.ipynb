```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import tkinter as tk
from tkinter import filedialog

# === JENDELA POP-UP UNTUK MEMILIH FILE SECARA OTOMATIS ===
print("⏳ Silakan pilih file 'dataset.csv' pada jendela pop-up yang muncul...")
root = tk.Tk()
root.withdraw() # Menyembunyikan jendela utama tkinter
root.attributes('-topmost', True) # Memaksa pop-up muncul di paling depan layar
file_path = filedialog.askopenfilename(
    title="Pilih File profil_mahasiswa.csv",
    filetypes=[("CSV Files", "*.csv")]
)

if not file_path:
    print("❌ Kamu membatalkan pilihan file. Proses dihentikan.")
else:
    print(f"📁 Berhasil memuat file dari: {file_path}")
    # Load dataset berdasarkan file yang kamu klik
    df = pd.read_csv(file_path)

    # Pilih fitur yang akan digunakan
    features = ["IPK", "Jumlah SKS", "Keaktifan Organisasi", "Keikutsertaan dalam Seminar/Pelatihan"]
    X = df[features]

    # Normalisasi data
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Menentukan jumlah cluster dengan metode Elbow
    inertia = []
    k_range = range(1, 10)
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42) 
        kmeans.fit(X_scaled)
        inertia.append(kmeans.inertia_)

    # Plot hasil Elbow Method
    plt.plot(k_range, inertia, marker='o')
    plt.xlabel('Jumlah Cluster (K)')
    plt.ylabel('Inertia')
    plt.title('Metode Elbow untuk Menentukan K Optimal')
    plt.show()

    # Menggunakan K optimal (misalnya, K=3)
    kmeans = KMeans(n_clusters=3, random_state=42)
    df['Cluster'] = kmeans.fit_predict(X_scaled)

    # Menampilkan hasil cluster
    print(df[['NIM', 'IPK', 'Jumlah SKS', 'Keaktifan Organisasi', 'Keikutsertaan dalam Seminar/Pelatihan', 'Cluster']])

    sns.scatterplot(x=df['IPK'], y=df['Jumlah SKS'], hue=df['Cluster'], palette='viridis')
    plt.title('Hasil Clustering Mahasiswa')
    plt.xlabel('IPK')
    plt.ylabel('Jumlah SKS')
    plt.show()
```
